```python
import matplotlib.pyplot as plt
import numpy as np

#DATA

#X
x=np.array([-3,-2,-1,0,1,2,3])

#Y
y=np.array([7.5,3,0.5,1,3,6,14])

# Calcular la regresión polinomial de segundo grado
grado = 2
coeficientes = np.polyfit(x, y, grado)
polinomio = np.poly1d(coeficientes)

# R Cuadrada
y_pred = polinomio(x)
rsquared = 1 - (np.sum((y - y_pred)**2) / ((len(y) - 1) * np.var(y, ddof=1)))

# Graficar
plt.scatter(x, y, color='b',label='Datos')
plt.plot(x, polinomio(x), color='r', linewidth = 2,label='Regresión polinomial de segundo grado')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión polinomial de segundo grado')
plt.legend()
plt.grid(linestyle = '--', linewidth = 0.5)
plt.show()

# Polinomio |  R²
print('\n')
print('Polinomio de la regresión polinomial de segundo grado:', polinomio)
print('\n')
print('R²:', rsquared)
```
